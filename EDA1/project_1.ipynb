#### 2. How do the number of accidents vary by day of the week

```python

accidents_by_day = data['week_day'].value_counts()
weekday_order = ['monday', 'tuesday', 'wednesday', 'thursday', 'friday', 'saturday', 'sunday']
accidents_by_day = accidents_by_day.reindex(weekday_order)
accidents_by_day.plot(kind='bar', title='Accidents by Day of the Week')
print(accidents_by_day)
```

#### 3. What are the peak hours for accidents during the day?


```python
data['hour_only'] = pd.to_datetime(data['hour'], format='%H:%M:%S').dt.hour
hourly_accidents = data['hour_only'].value_counts().sort_index()
peak_hours = hourly_accidents[hourly_accidents == hourly_accidents.max()]
print("Hourly Accident Data:")
print(hourly_accidents)
print("\nPeak Hours (with maximum accidents):")
print(peak_hours)
plt.figure(figsize=(10, 6))
hourly_accidents.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title("Accidents by Hour", fontsize=16)
plt.xlabel("Hour of the Day", fontsize=14)
plt.ylabel("Number of Accidents", fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

#### 4. Which cities report the highest number of accidents?


```python
city_accident_counts = data['city'].value_counts()
top_cities = city_accident_counts.head(10)
top_cities.plot(kind='bar', title='Top 10 Cities with the Most Accidents', color='green', edgecolor='black')
plt.xlabel('City')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45, ha='right')
plt.show()
print(city_accident_counts)
```

#### 5. Which states have the highest and lowest accident frequencies?


```python
state_accident_counts = data['state'].value_counts()
highest_state = state_accident_counts.idxmax()
lowest_state = state_accident_counts.idxmin()
print(f"The state with the highest number of accidents: {highest_state} ({state_accident_counts.max()} accidents)")
print(f"The state with the lowest number of accidents: {lowest_state} ({state_accident_counts.min()} accidents)")
state_accident_counts.plot(kind='bar', title='Accident Frequencies by State', color='blue', edgecolor='black')
plt.xlabel('State')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45, ha='right')
plt.show()
print(state_accident_counts)
```

#### 7.How does the severity of accidents vary byroad type?

#### 8. Are accidents more common under certain weather conditions?

#### 9.What is the relationship between the number of vehicles involved and accident severity?


```python
import pandas as pd
import matplotlib.pyplot as plt
file_path = "C:\Data Science\Project\Accidents.csv"
data = pd.read_csv(file_path)
print("Dataset Head:")
print(data.head())
severity_by_vehicles = data.groupby('vehicles_involved')['victims_condition'].value_counts().unstack()
print("Relationship Between Vehicles Involved and Accident Severity:")
print(severity_by_vehicles)
severity_by_vehicles.plot(kind='bar',stacked=True,figsize=(12,7),colormap ='tab10')
plt.title("Relationship Between Vechicles Involved and Accident Severity",fontsize =16)
plt.xlabel("Number of Vehicles involved",fontsize = 14)
plt.ylabel("Number of Accidents",fontsize=14)
plt.legend(title="Accident Severity",loc='upper left' )
plt.grid(axis='y',linestyle='--',alpha = 0.7)
plt.show()
```

#### 10.Which regions have the highest number of severely injured victims?
